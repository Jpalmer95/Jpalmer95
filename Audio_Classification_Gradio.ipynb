```python
from transformers import pipeline

#model_id = "sanchit-gandhi/distilhubert-finetuned-gtzan"
model_id = "anthony-wss/distilhubert-finetuned-gtzan"
pipe = pipeline("audio-classification", model=model_id)
```

```python
# Install the required module
!pip install gradio
```

```python

# Import the required modules
import gradio as gr

# Define the function to be used by Gradio
def classify_audio(filepath):
    preds = pipe(filepath)
    outputs = {}
    for p in preds:
        outputs[p["label"]] = p["score"]
    return outputs
    # Return the predicted label
    pass

# Create the Gradio interface
demo = gr.Interface(
    fn=classify_audio,  # Use the defined function
    inputs=gr.Audio(type="filepath"),  # Use filepath as input
    outputs=("label")  # Use a label to display the output
)

# Launch the Gradio interface
demo.launch(debug=True)
```
